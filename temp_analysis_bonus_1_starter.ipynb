### Compare June and December data across all years 

```python
from scipy import stats
```

```python
csvpath=os.path.join('Resources', 'hawaii_measurements.csv')
df = pd.read_csv(csvpath)
# Convert the date column format from string to datetime
df['date']=pd.to_datetime(df['date'])
df.head()
```

```python
# Filter data for desired months (June)
jun_df = df.loc[df["date"].dt.month== 6, :]
jun_df
```

```python
# Filter data for desired months (December)
dec_df = df.loc[df["date"].dt.month== 12, :]
dec_df
```

```python
# Identify the average temperature for June
avejun= jun_df["tobs"].mean()
print("June Average temperature is : ", avejun)
```

```python
# Identify the average temperature for December
avedec= dec_df["tobs"].mean()
print("June Average temperature is : ", avedec)
```

```python
# Create collections of temperature data
junvalues= jun_df.loc[:, 'tobs']
decvalues= dec_df.loc[:,'tobs']
```

```python
# Run paired t-test
t=stats.ttest_ind(junvalues, decvalues)
t
```

### Analysis

```python
print("Los resultados son estadísticamente significativos con valor de P= ",t.pvalue)

print("Pvalue is less than 0.05, means hypotesis may be true")
```
